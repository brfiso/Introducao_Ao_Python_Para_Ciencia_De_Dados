### Visualizando a estrutura da tabela
A função .head() exibe as primeiras linhas da lista


1. Pregnancies: To express the Number of pregnancies
* Número de Gravidezes. Refere-se à quantidade de vezes que a pessoa esteve grávida.

2. Glucose: To express the Glucose level in blood
* Glicose. Indica o nível de glicose (açúcar) presente no sangue. Geralmente é medido após um período de jejum.

3. BloodPressure: To express the Blood pressure measurement
 Pressão Sanguínea. Representa a pressão que o sangue exerce contra as paredes das artérias. É geralmente expressa por dois números: a pressão sistólica (quando o coração se contrai) e a diastólica (quando o coração relaxa).

4. SkinThickness: To express the thickness of the skin
* Espessura da Pele. Medida da espessura da dobra cutânea, geralmente na região do tríceps. Em alguns contextos, pode ser usada como uma estimativa da gordura corporal.

5. Insulin: To express the Insulin level in blood
* Insulina. Indica o nível de insulina no sangue. A insulina é um hormônio produzido pelo pâncreas que ajuda a glicose a entrar nas células para ser usada como energia.

6. BMI: To express the Body mass index
* Índice de Massa Corporal (IMC). É uma medida que relaciona o peso de uma pessoa com a sua altura. É usado para classificar se uma pessoa está abaixo do peso, com peso saudável, com sobrepeso ou obesa.

7.  DiabetesPedigreeFunction: To express the Diabetes percentage
* Função de Hereditariedade do Diabetes. Representa uma pontuação que leva em consideração o histórico familiar de diabetes e a predisposição genética da pessoa para desenvolver a doença.

8. Age: To express the age
*  Idade. Refere-se à idade da pessoa em anos.

9. Outcome: To express the final result 1 is Yes and 0 is No
* Resultado. Indica o resultado final, onde 1 significa que a pessoa foi diagnosticada com diabetes e 0 significa que não foi diagnosticada com diabetes.

#### Correlação entre as variáveis
.corr() é usado para criar uma matriz de correção
.heatmap() cria um gráfico de calor



```python
from sklearn.preprocessing import LabelEncoder
df['Ethnicity'] = LabelEncoder().fit_transform(df['Ethnicity'])
df['Alcohol_Consumption'] = LabelEncoder().fit_transform(df['Alcohol_Consumption'])

sns.heatmap(df[['Age', 'Ethnicity', 'BMI', 'Blood_Pressure_Systolic', 'Cholesterol_LDL', 'Cholesterol_HDL', 'Family_History_of_Diabetes', 'Alcohol_Consumption', 'Blood_Pressure_Systolic']].corr())
```

```python
for col in ['Alcohol_Consumption']:
    df[col] = LabelEncoder().fit_transform(df[col])
    
sns.heatmap(df[['Age', 'Alcohol_Consumption', 'Cholesterol_HDL', 'Cholesterol_LDL', 'Blood_Pressure_Systolic']].corr(), annot=True)
```
